```python
import os
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

### File Load

```python
root_folder = "D:/hyundai/dataset/data/original/2024"
csv_root_folder = "D:/hyundai/dataset/data/roi"

file_list = []
csv_list = []

for folder_name, subfolders, filenames in os.walk(root_folder):
    folder_list = []
    for filename in filenames:
        if filename.lower().endswith('.jpg'):
            file_list.append(os.path.join(folder_name, filename))
    
    # if folder_list:
        # file_list.append(folder_list)

for folder_name, _, filenames in os.walk(csv_root_folder):
    for filename in filenames:
        if filename.lower().endswith('.csv'):
            csv_list.append(os.path.join(folder_name, filename))

a = len(file_list)
b = len(csv_list)
print(a)
print(b)
```

```python
folder_names = set()

for file_path in file_list:
    for f_name in file_path:
        dir_path = os.path.dirname(f_name)
        folder_name = os.path.basename(dir_path)
        folder_names.add(folder_name)

print(len(folder_names))
```

```python

for files in file_list:
    for file in files:
        image = Image.open(file)

        parts = file.split('\\')
        key_part = parts[-2]  
        key = key_part.split('~')[0]

        roi_list = dfs[key]

        for row_idx in range(len(roi_list)):
            for col_start in range(0, len(roi_list.columns), 8):
                coords = [(roi_list.iloc[row_idx, col_start + i], roi_list.iloc[row_idx, col_start + i + 1]) for i in range(0, 8, 2)]
                print(coords)

                if not all(x == 0 for coord in coords for x in coord):
                    left = min(coords, key=lambda x: x[0])[0]
                    top = min(coords, key=lambda x: x[1])[1]
                    right = max(coords, key=lambda x: x[0])[0]
                    bottom = max(coords, key=lambda x: x[1])[1]

                    roi_image = image.crop((left, top, right, bottom))

                    crop_save_path = os.path.join("crop", key, os.path.basename(file))
                    os.makedirs(os.path.dirname(crop_save_path), exist_ok=True)
                    roi_image.save(crop_save_path)

                    break

                else:
                    continue

            break

    break
```
